```python
def stringsplit(string,length):
    brokenString=[]
    while len(string)>=length:
        brokenString.append(string[0:length])
        string=string[length:]
    if len(string)==0:
        return brokenString
    else:
        for a in range(length-len(string)):
            string=string+chr(65+random.randint(0,25))
        brokenString.append(string)
        return brokenString
```
